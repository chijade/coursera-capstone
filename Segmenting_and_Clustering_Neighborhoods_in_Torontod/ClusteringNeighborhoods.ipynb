```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```
wiki_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

```

```
page = requests.get(wiki_url)
```

```
data = page.text
```

```
soup = BeautifulSoup(data)
```

```
print(soup.prettify())
```

```
wiki_table = soup.find('table')
wiki_table
```

```
table_row = wiki_table.find_all('tr')
table_row
```

```
dataframe = []
for row in table_row:
    table_data = row.find_all('td')
    data = [i.text.rstrip() for i in table_data]
#     print(data)
    dataframe.append(data)

dataframe 
del dataframe[0]
dataframe
```

```
labels = ['Postcode', 'Borough', 'Neighborhood']
```

```
wiki_table_df = pd.DataFrame.from_records(dataframe, columns=labels)
```

```
wiki_table_df = wiki_table_df[wiki_table_df.Borough != 'Not assigned']

```

```
wiki_table_df
```

```
wiki_table_df.shape
```

```
geo_url ='http://cocl.us/Geospatial_data'
wiki_table_df_geo_coor = requests.get(geo_url)
wiki_table_df_geo_coor = pd.read_csv(geo_url)
wiki_table_df_geo_coor.head()

```

```
df_toronto = pd.merge(wiki_table_df, wiki_table_df_geo_coor, how='left', left_on = 'Postcode', right_on = 'Postal Code')
# remove the "Postal Code" column
df_toronto.drop("Postal Code", axis=1, inplace=True)
df_toronto.head()
```

```
from geopy.geocoders import Nominatim
geolocator = Nominatim()
location = geolocator.geocode("Toronto")
print(location.address)
print('The geograpical coordinate of Toronto city are {}, {}.'.format(location.latitude, location.longitude))
```

```
import folium
```

```
#create a map of the whole Toronto
map_toronto = folium.Map([location.latitude, location.longitude], zoom_start=12)
map_toronto
```

```
df_specific = df_toronto[df_toronto['Borough'].str.contains('Toronto',regex=False)]
df_specific
```

```
#map of all the boroughs following Toronto
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df_specific['Latitude'],df_specific['Longitude'],df_specific['Borough'],df_specific['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```
#Using KMeans clustering for the clsutering of the neighbourhoods
from sklearn.cluster import KMeans
k=5
toronto_clustering = df_specific.drop(['Postcode','Borough','Neighborhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
```

```
df_specific
```

```
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```
df_specific.insert(0, 'Cluster Labels', kmeans.labels_)
```

```
# create map of clusters
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=12)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighborhood, cluster in zip(df_specific['Latitude'], df_specific['Longitude'], df_specific['Neighborhood'],df_specific['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
